```
import pandas as pd
import plotly.express as px
```

```
df = pd.read_csv("final_table.csv")
df
```

**Stats**
Calculate stats for age, hours per day listening to music, frequency of different genres, anxiety levels, depression levels, insomnia levels and OCD **levels**

**Ages that listen to most hours per day and their results**

```
df2 =round(df[['Age',"Hours_per_day","Anxiety","Depression","Insomnia","OCD"]][(df["Age"] >= 18)&(df["Age"] <= 62)&(df["Age"] != 55)&(df["Age"] != 54)].groupby("Age",as_index=False).mean(),2)
df2
```

```
px.line(x=df2['Age'],y=df2['OCD'])
```

It is important to notice, that when the average hours per day that people listen to music tends to lower its levels as the people gets older.
The levels of anxiety tends to be pretty high on ages from 20 to 30, dropping its level after 45 years old, similar to Depression
The levels of insomnia tends to be higher after 35
OCD levels are being reduced as the age increases, but after 50 the general levels of OCD are higher.

```
df_3 = round(df[["Age","Hours_per_day","Anxiety","Depression","Insomnia","OCD"]].corr(),1)
df_3
```

The strongest correlation is found between anxiety and depression, that means
that people that feels anxiety are more likely to feel depressed over time.
The hours per day that people listen to music does not seems to have a relevant effect on peoples perception of Anxiety, Depression, Insomnia or OCD.

```
df_4 = round(df[["Fav_genre",'Age',"Hours_per_day","Anxiety","Depression","Insomnia","OCD"]][(df["Age"] >= 18)&(df["Age"] <= 62)].groupby("Fav_genre").mean().sort_values(by=['Anxiety'],ascending=False).head(20),1)
df_4
```

The people that reports Lofi as their fav_genre tends to have higher levels of Anxiety and Depression, also the average age of this people is 24.8 years old, that supports the previous analysis on the levels of anxiety and depression on the mid 20s.

```
df_5 = round(df[["Fav_genre","Hours_per_day","Anxiety"]][(df["Age"] >= 18)&(df["Age"] <= 62)].groupby("Fav_genre").mean().sort_values(by=['Anxiety'],ascending=False).head(20),2).corr()
df_5
```

Analyzing the data from ages between 18 and 62 years old, we can see that the correlation between hours_per_day and anxiety tends to increase its value.

```
df_6 = round(df[["Frequency_Lofi","Age","Hours_per_day","Anxiety","Depression","Insomnia","OCD"]][(df["Age"] >= 18)&(df["Age"] <= 62)].groupby("Frequency_Lofi").mean(),2)
df_6
```

Analyzing the worst 3 genres in general in terms of levels of depression, anxiety, insomnia, OCD, etc. The levels of Anxiety tends to be higher when the people listen to this genre "sometimes", that means 3.78 hours a day.

```
df_7 = round(df[["Frequency_Rap","Age","Hours_per_day","Anxiety","Depression","Insomnia","OCD"]][(df["Age"] >= 18)&(df["Age"] <= 62)].groupby("Frequency_Rap").mean(),2)
df_7
```

In the case of Rap, the people that presents the higher levels of anxiety listen to them rarely, that means less than 3.3 hours a day.

```
df_8 =round(df[["Frequency_EDM","Age","Hours_per_day","Anxiety","Depression","Insomnia","OCD"]][(df["Age"] >= 18)&(df["Age"] <= 62)].groupby("Frequency_EDM").mean(),1)
df_8
```

In the case of EDM it presents the highest levels of insomnia for people that listen to it very frequently, that means more than 4.1 hours a day.
