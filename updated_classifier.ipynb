# Correlates anxiety with other columns

```python
# computes the correlation coefficient between the "Anxiety"
# column and each other column in the DataFrame df
# calculating two different measures of association (Cramér's V and Theil's U) between the
# 'Anxiety' column and two other columns ('Hours_per_day' and 'While_working') in the DataFrame
# df, and then printing out the results. This allows you to assess the degree of association or
# correlation between these categorical variables in your dataset.

import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Example data (replace this with your actual data)

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# stores the result of calculating Cramér's V between the 'Anxiety' column and the 'Hours_per_day'
cramers_v_value = cramers_v(df['Anxiety'], df['Hours_per_day'])
cramers_v_value_w = cramers_v(df['Anxiety'], df['While_working'])

print("Cramér's V for Anxiety vs Hours per day:", cramers_v_value)
print("Cramér's V for Anxiety vs While Working:", cramers_v_value_w)
```

**Anxiety vs Hours per day**:  This value is quite low, indicating a very weak association between the variables "Anxiety" and "Hours per day". Essentially, this means that the amount of time spent listening to music per day has a very weak relationship with reported anxiety levels in your dataset. This low value suggests that variations in the hours per day spent listening to music **do not** strongly correlate with changes in anxiety levels.

**Anxiety vs While Working**: This value is also relatively low, though slightly higher than the first value, indicating a weak association between the variables "Anxiety" and "While Working". This means that whether a person listens to music while working has a weak relationship with their reported anxiety levels. While the association is slightly stronger than that between "Anxiety" and "Hours per day", it is still weak and suggests that listening to music while working **does not** strongly correlate with changes in anxiety levels.

# Drop `BPM_range` column

```python
# Function to convert range to mean value
def range_to_mean(value):
    min_val, max_val = map(int, value.split(' - '))
    return (min_val + max_val) / 2

# Apply the function to convert the range strings to numerical values
df['BPM_mean'] = df['BPM_range'].apply(range_to_mean)

# Calculate the median of these numerical values
median_bpm = df['BPM_mean'].median()

# Replace original range values with the median
df['BPM_range'] = median_bpm

# Drop the temporary 'BPM_mean' column
df = df.drop(columns=['BPM_mean'])

print("DataFrame after replacing BPM_range with median value:")
df.head(20)
```
